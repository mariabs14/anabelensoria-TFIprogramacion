Crear un red de ocho nodos que muestre una relación direccionada, de manera que cada nodo sólo pueda tener comunicación directa con otros 2 nodos.

```python
 # Importamos las librerias Networkx  y Matplotlib
import networkx as nx
import matplotlib.pyplot as plt

def RED_NODOS():
    # Se crear el grafico en anillo
    G = nx.DiGraph()

    # Definir los 8 nodos
    G.add_nodes_from(range(1, 9))
    print("El listado de Nodos es: {}".format(list(G.nodes)))

    # Generar la conexión: cada nodo se comunica a un solo destino
    # y recibe de un solo origen, sumando exactamente 2 conexiones por nodo.
    aristas = []
    for i in range(8):
        nodo_origen = i + 1
        # El módulo asegura que el nodo 8 se conecte al nodo 1 para cerrar el anillo
        nodo_destino = ((i + 1) % 8) + 1

        aristas.append((nodo_origen, nodo_destino))

    G.add_edges_from(aristas)

    # Configuración gráfica para representar los 8 nodos en una circunsferencia
    plt.figure(figsize=(7, 7))
    pos = nx.circular_layout(G)

    # Dibujar los nodos
    nx.draw_networkx_nodes(G, pos, node_color='#ffb3b3', node_size=1000, edgecolors='black')
    nx.draw_networkx_labels(G, pos, font_size=12, font_weight='bold')

    # Dibujar los enlaces con curvatura en las flechas
    nx.draw_networkx_edges(
        G, pos,
        arrows=True,
        arrowsize=10,
        edge_color='#6c757d',
        width=2,
        connectionstyle='arc3,rad=0.15'
    )

    # Títulos
    plt.title("Redes y Comunicación\n(RED DE 8 NODOS)", fontsize=15, pad=15)
    plt.axis('off')
    plt.show()

# Final

RED_NODOS()
```
